```python
# which 10 companies make the most money?
SELECT
    dc.company as company,
    SUM(ftt.fare) as total_fare
FROM
    chicago_taxi_db.fact_taxi_trips ftt
INNER JOIN
    chicago_taxi_db.dim_company dc
ON
    ftt.company_id = dc.company_id
GROUP BY
    company
ORDER BY
    total_fare DESC
LIMIT 10;
```

```python
# top 10 pickup community areas with the most rides
SELECT
    dca.community_area as community_area,
	COUNT(ftt.trip_id) as rides
FROM
    chicago_taxi_db.fact_taxi_trips ftt
INNER JOIN
    chicago_taxi_db.dim_community_areas dca
ON
    dca.area_code = ftt.pickup_community_area_id
GROUP BY
    community_area
ORDER BY
    rides DESC
LIMIT 10;
```

```python
# count taxi rides per day of week
SELECT
    CASE dd.day_of_week
        WHEN 1 THEN 'Monday'
        WHEN 2 THEN 'Tuesday'
        WHEN 3 THEN 'Wednesday'
        WHEN 4 THEN 'Thursday'
        WHEN 5 THEN 'Friday'
        WHEN 6 THEN 'Saturday'
        WHEN 7 THEN 'Sunday'
    END
    AS weekday_name,
    COUNT(ftt.trip_id) AS taxi_rides_per_day
FROM fact_taxi_trips ftt
JOIN dim_date dd
    ON ftt.trip_start_timestamp = dd.date
GROUP BY dd.day_of_week
ORDER BY dd.day_of_week;
```
