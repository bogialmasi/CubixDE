```python
# which 10 companies make the most money?
SELECT
    dc.company as company,
    SUM(ftt.fare) as total_fare
FROM
    chicago_taxi_db.fact_taxi_trips ftt
INNER JOIN
    chicago_taxi_db.dim_company dc
ON
    ftt.company_id = dc.company_id
GROUP BY
    company
ORDER BY
    total_fare DESC
LIMIT 10;
```

```python
# top 10 pickup community areas with the most rides
SELECT
    dca.community_area as community_area,
	COUNT(ftt.trip_id) as rides
FROM
    chicago_taxi_db.fact_taxi_trips ftt
INNER JOIN
    chicago_taxi_db.dim_community_areas dca
ON
    dca.area_code = ftt.pickup_community_area_id
GROUP BY
    community_area
ORDER BY
    rides DESC
LIMIT 10;
```
