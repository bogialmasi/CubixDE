```python
import json
```

```python
filepath = r'D:\Cubix DE II\Code\data\json_scraping\spotify_playlist.json'

with open(filepath, 'r') as file:
	data = json.load(file)
```

```python
# name of the 31st track
print(data['contents']['items'][32]['name'])
```

```python
# total playcount for all tracks
playCount_sum = 0

for track in data['contents']['items']:
	playCount_sum += track['playCount'] # add to the total

print(playCount_sum)
```

```python
# song with the lowest playcount
playCount_min = data['contents']['items'][0]

for track in data['contents']['items']:
	if track['playCount'] < playCount_min['playCount']:
		playCount_min = track

song_name = playCount_min["name"]
print(f'The song with the lowest playcount is {song_name}')
```
