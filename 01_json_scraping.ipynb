## JSON fájl beolvasása

```python
# json fájl útvonala
filepath = r'D:\Cubix DE II\Code\data\json_scraping\spotify_playlist.json'

with open(filepath, 'r') as file:
	data = json.load(file) # data tartalma a JSON tartalma
```

## számok adatainak kiírása

```python
# 0. zene címe
data['contents']['items'][0]['name']
```

```python
# minden előadó neve
for track in data['contents']['items']:
	for artist in track['artists']:
		print(artist['name'])
	print()
```

```python
# melyik számot hányszor játszották le
for track in data['contents']['items']:
	print(track['name'], ',', track['playCount'])
```
