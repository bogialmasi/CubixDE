### egy nap tripjei

```python
# éjféltől éjfélig, max 30 000 sor
url = (f"https://data.cityofchicago.org/resource/ajtu-isnz.json?"
	   f"$where=trip_start_timestamp >= '2024-02-16T00:00:00' "
	   f"AND trip_start_timestamp <= '2024-02-16T23:59:59'"
	   f"&$limit=30000")
response = requests.get(url)
data = response.json()
len(data)
```

### egy kiválasztott nap tripjei

```python
# mai dátumhoz képest 2 hónappal korábban
current_datetime = datetime.now() - relativedelta(months = 2) # datetime típus
formatted_datetime = current_datetime.strftime("%Y-%m-%d") # string from time formázás
formatted_datetime
```

```python
# éjféltől éjfélig, max 30 000 sor
url = (f"https://data.cityofchicago.org/resource/ajtu-isnz.json?"
	   f"$where=trip_start_timestamp >= '{formatted_datetime}T00:00:00' "
	   f"AND trip_start_timestamp <= '{formatted_datetime}T23:59:59'"
	   f"&$limit=30000")
response = requests.get(url)
data = response.json()
len(data)
```
